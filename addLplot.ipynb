```python
#import cv2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('c_kneebone_fix_total_add_prob_addL.csv')
df = df.reset_index()
df['right_medial_prob'] = 0
df['right_f_prob'] = 0
df['left_medial_prob'] = 0
df['left_f_prob'] = 0
df['letter_R_prob'] = 0
df['letter_L_prob'] = 0
df = df.shift(periods=6, axis="columns")
df = df.drop(["level_0",'level_1','level_2','level_3','level_4','level_5'], axis = 1)
df = df.sort_values(by = 'file_name', ascending = True)
df.to_csv("predict_addL.csv", mode='w')
df
```

```python
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 11.7,8.27
```

```python
fig, ax = plt.subplots(ncols=2)
sns.boxplot("letter_R_prob", data=df, ax=ax[0])
sns.histplot(df['letter_R_prob'], ax=ax[1])
```

```python
fig, ax = plt.subplots(ncols=2)
sns.boxplot("letter_L_prob", data=df, ax=ax[0])
sns.histplot(df['letter_L_prob'], ax=ax[1])
```
