```python
import pandas as pd
```

```python
new = pd.read_csv('c_kneebone_fix_total_add_prob_new.csv')
new.head()
```

```python
new = new.reset_index()
new['right_medial_prob'] = 0
new['right_f_prob'] = 0
new['left_medial_prob'] = 0
new['left_f_prob'] = 0
new['letter_R_prob'] = 0
new['letter_L_prob'] = 0
```

```python
new = new.shift(periods=6, axis="columns")
new
```

```python
new = new.drop(["level_0",'level_1','level_2','level_3','level_4','level_5'], axis = 1)
new.to_csv("c_kneebone_fix_total_add_prob_new_make.csv", mode='w')
new
```

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
import seaborn as sns
# figure size in inches
rcParams['figure.figsize'] = 11.7,8.27
```

```python
fig, ax = plt.subplots(ncols=2)
sns.boxplot("letter_R_prob", data=new, ax=ax[0])
sns.histplot(new['letter_R_prob'], ax=ax[1])
```

```python
fig, ax = plt.subplots(ncols=2)
sns.boxplot("letter_L_prob", data=new, ax=ax[0])
sns.histplot(new['letter_L_prob'], ax=ax[1])
```
